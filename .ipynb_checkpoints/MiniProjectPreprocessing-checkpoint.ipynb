```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# ----------------------------
# 1 Load dataset
# ----------------------------
df = pd.read_csv("dataset.csv")

print("Original Shape:", df.shape)

# ----------------------------
# 2 Remove duplicates
# ----------------------------
df = df.drop_duplicates()
print(df.head(5))


features = ["EAR", "MAR", "Head_Tilt"]

# drop rows where feature values missing
df = df.dropna(subset=features)

# drop rows where label missing
df = df.dropna(subset=["Label"])

# keep only valid labels
df = df[df["Label"].isin([0,1])]

print("After basic cleaning:", df.shape)
# remove mislabeled samples

df = df[~((df["EAR"] > 0.30) & (df["Label"] == 1))]
df = df[~((df["EAR"] < 0.18) & (df["Label"] == 0))]
df = df[~((df["MAR"] > 0.7) & (df["Label"] == 0))]
# ----------------------------
# 6 Outlier Handling (Robust IQR)
# Instead of deleting aggressively,
# we clip extreme values
# ----------------------------
# remove mislabeled samples

# Features
features = ["EAR", "MAR","Head_Tilt"]

# ----------------------------
# 1 Handle EAR and MAR normally (IQR clipping)
# ----------------------------
# EAR limits
df["EAR"] = df["EAR"].clip(0.1, 0.4)

# MAR limits
df["MAR"] = df["MAR"].clip(0, 0.35)
df = df[df["MAR"] > 0.005]

# Head tilt limits
df["Head_Tilt"] = df["Head_Tilt"].clip(-15, 15)

print("After outlier handling:", df.shape)
print("After outlier handling:", df.shape)

# ----------------------------
# 7 Feature / Label split
# ----------------------------
X = df[features]
y = df["Label"]

# ----------------------------
# 8 Feature Scaling
# ----------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ----------------------------
# 9 Train Test Split
# ----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

print("Training Samples:", X_train.shape)
print("Testing Samples:", X_test.shape)

#plotting the ear,mar and head_tilt values

fig, axes = plt.subplots(1,3, figsize=(15,4))

sns.boxplot(y=df["EAR"], ax=axes[0])
axes[0].set_title("EAR")

sns.boxplot(y=df["MAR"], ax=axes[1])
axes[1].set_title("MAR")

sns.boxplot(y=df["Head_Tilt"], ax=axes[2])
axes[2].set_title("Head Tilt")

plt.tight_layout()
plt.show()

# ----------------------------
# 10 Save cleaned dataset
# ----------------------------
cleaned_df = pd.DataFrame(X_scaled, columns=features)
cleaned_df["Label"] = y.values

cleaned_df.to_csv("cleaned_dataset.csv", index=False)

print("Cleaned dataset saved successfully.")
```
